```python
import clustering
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn_extra.cluster import KMedoids
```

```python
file_name = "data/dataset.csv"
clusterator = clustering.Clusterator(file_name=file_name)
```

```python
# clusterator.clusterize("price", model=KMeans(3))
clusterator.clusterize(columns=["latitude", "longitude", "price"], model=DBSCAN(eps=0.101, min_samples=6))
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15,10))
for i, ax in enumerate(axs.flatten()):
    model = KMeans(i+2)
    scatter = clusterator.clusterize(columns=["latitude", "longitude"], model=model, ax=ax)
plt.tight_layout()
plt.show()
```
