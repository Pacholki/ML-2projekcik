```python
import clustering
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from hdbscan import HDBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn import mixture
from sklearn_extra.cluster import KMedoids
```

```python
file_name = "data/dataset.csv"
clusterator = clustering.Clusterator(file_name=file_name)
```

# KMeans sensitivity to outliers

```python
fig, axs = plt.subplots(2, 3, figsize=(15,10))
for i, ax in enumerate(axs.flatten()):
    model = KMeans(i+2)
    clusterator.clusterize(columns=["latitude", "longitude", "price"], model=model)
    clusterator.plot(ax=ax)
plt.tight_layout()
plt.show()
```

```python
clusterator.clusterize("price", model=KMeans(6))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["latitude", "longitude", "price"], model=DBSCAN(eps=0.101, min_samples=6))
clusterator.plot()
```

# Normalized column values

```python
fig, axs = plt.subplots(1, 2, figsize=(20,4))
clusterator.boxplotter(df=clusterator.df, column="price", ax=axs[0])
clusterator.boxplotter(column="price", ax=axs[1])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20,4))
clusterator.boxplotter(df=clusterator.df, column="longitude", ax=axs[0])
clusterator.boxplotter(column="longitude", ax=axs[1])
```

```python
clusterator.clusterize(columns=["latitude", "longitude", "reviews_per_month"], model=DBSCAN(eps=0.081, min_samples=6))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["number_of_reviews", "longitude", "latitude"], model=DBSCAN(eps=0.073, min_samples=6))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["minimum_nights", "longitude", "latitude"], model=DBSCAN(eps=0.2, min_samples=6))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["price", "reviews_per_month"], model=AgglomerativeClustering(n_clusters=5, linkage="ward"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["calculated_host_listings_count", "reviews_per_month"], model=AgglomerativeClustering(n_clusters=5, linkage="average"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["longitude", "latitude"], model=mixture.GaussianMixture(n_components=5, covariance_type="full"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["longitude", "latitude", "price"], model=mixture.GaussianMixture(n_components=5, covariance_type="diag"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["longitude", "latitude", "price"], model=mixture.GaussianMixture(n_components=6, covariance_type="spherical"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["longitude", "latitude", "popularity"], model=mixture.GaussianMixture(n_components=6, covariance_type="full"))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["popularity", "price"], model=DBSCAN(eps=0.04, min_samples=4))
clusterator.plot()
```

```python
clusterator.clusterize(columns=["minimum_nights", "longitude", "latitude"], model=HDBSCAN(cluster_selection_epsilon=0.2, min_samples=6))
clusterator.plot()
```
