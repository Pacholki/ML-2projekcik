```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/dataset.csv")
```

```python
clustering = ["neighbourhood_group", "price", "number_of_reviews", "reviews_per_month", "beds", "price"]

fig, axs = plt.subplots(2, 3, figsize=(15,10))
for i, ax in enumerate(axs.flatten()):
    if i == 0:
        continue
    df.plot(x="longitude", y="latitude", kind="scatter", c=clustering[i], ax=ax)
    # sns.scatterplot(data=df, x="longitude", y="latitude", hue=clustering[i], ax=ax)
    ax.set_title(f"Clustering by {clustering[i]}")

sns.scatterplot(data=df, x="longitude", y="latitude", hue=clustering[0], ax=axs.flatten()[0])
axs.flatten()[0].set_title(f"Clustering by {clustering[0]}")


plt.tight_layout()
plt.show()
```
